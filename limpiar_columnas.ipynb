```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# reading the file
df_raw_data= pd.read_csv('HR_RAW_DATA.csv', index_col=0)
df_raw_data.head(3)
```

* columna 12

```python
# Remove  employeecount column
df_raw_data.drop("employeecount", axis = 1, inplace = True)
df_raw_data.head(1)
```

* columna 14 

```python
# Rename
df_raw_data = df_raw_data.rename(columns={'EnvironmentSatisfaction': 'Environment_Satisfaction'})

# Convert to String
df_raw_data['Environment_Satisfaction'] = df_raw_data['Environment_Satisfaction'].astype(str)

#Changing values
for i, val in enumerate(df_raw_data['Environment_Satisfaction']):
    if len(val) == 2:
        # Add a dot in the middle
       df_raw_data.loc[i, 'Environment_Satisfaction'] = val[0] + '.' + val[1]
    elif len(val) == 1:
        # Add a zero at the end:
       df_raw_data.loc[i, 'Environment_Satisfaction'] = val + '.0'
       
#Convert to float
df_raw_data['Environment_Satisfaction'] = df_raw_data['Environment_Satisfaction'].astype(float)

# Checking valid values within the range of (1,5) in the column
df_raw_data[(df_raw_data['Environment_Satisfaction'] >= 1) & (df_raw_data['Environment_Satisfaction'] <= 5)].head(5)
```

* columna 15

```python
df_raw_data['Environment_Satisfaction'].unique()
```

```python
# Gender

# Convert to String
df_raw_data['Gender'] = df_raw_data['Gender'].astype(str)

# Changing values
df_raw_data['Gender'] = df_raw_data['Gender'].replace({'0': 'M', '1': 'F'})
```

```python
df_raw_data['Gender'].unique()
```

* columna 16

```python
# Rename
df_raw_data = df_raw_data.rename(columns={'HourlyRate': 'Hourly_Rate'})

# Removing 'Not Available'
df_raw_data['Hourly_Rate'] = df_raw_data['Hourly_Rate'].replace({'Not Available': '99999'})

# Convert to Integer
df_raw_data['Hourly_Rate'] = df_raw_data['Hourly_Rate'].astype(int)
```

* columna 17

```python
# Rename
df_raw_data = df_raw_data.rename(columns={'JobInvolvement': 'Job_Involvement'})

#Convert to Integer
df_raw_data['Job_Involvement'] = df_raw_data['Job_Involvement'].astype(int)

# Checking valid values within the range of (1,5) in the column.
df_raw_data[(df_raw_data['Job_Involvement'] >= 1) & (df_raw_data['Job_Involvement'] <= 5)].head(5)		
```

columna 18

```python
# Rename
df_raw_data = df_raw_data.rename(columns={'JobLevel': 'Job_Level'})

## Checking valid values within the range of (1,5) in the column.
df_raw_data[(df_raw_data['Job_Level'] >= 1) & (df_raw_data['Job_Level'] <= 5)].head(5)	
```

* columna 19

```python
# Rename
df_raw_data = df_raw_data.rename(columns={'JobRole': 'Job_Role'})

# Sorting column values
df_raw_data['Job_Role'] = df_raw_data['Job_Role'].str.title()
```

columna 20

```python
#Rename
df_raw_data = df_raw_data.rename(columns={'JobSatisfaction': 'Job_Satisfaction'})

# Checking valid values within the range of (1,5) in the column.
df_raw_data[(df_raw_data['Job_Satisfaction'] >= 1) & (df_raw_data['Job_Satisfaction'] <= 5)].head(5)
```

columna 21

```python
# Rename
df_raw_data = df_raw_data.rename(columns={'MaritalStatus': 'Marital_Status'})

# Sorting column values
df_raw_data['Marital_Status'] = df_raw_data['Marital_Status'].str.title()

# Ortografic errors
df_raw_data['Marital_Status']= df_raw_data['Marital_Status'].replace({'Marreid':'Married'})
```

* columna 22

```python
# Rename
df_raw_data = df_raw_data.rename(columns={'MonthlyIncome': 'Monthly_Income'})

# Replace (.)
df_raw_data['Monthly_Income'] = df_raw_data['Monthly_Income'].str.replace(",", ".")

# Convert to float
df_raw_data['Monthly_Income'] = df_raw_data['Monthly_Income'].astype(float)
```
