```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('HR_CLEAN_DATA.csv')
df.head(3)
```

```python
valores_unicos_Monthly_Income = df['Monthly_Income'].unique()
print("Valores únicos en la columna 'Monthly_Income':")
print(valores_unicos_Monthly_Income)
```

```python
frecuencia_marital_status = df['Marital_Status'].value_counts()
print("\nFrecuencia de cada valor único en la columna 'Marital_Status':")
print(frecuencia_marital_status)
```

```python
nulos_Marital_Status_count = df['Marital_Status'].isnull().sum()
print(f"Número de valores nulos en 'Marital_Status': {nulos_Marital_Status_count}")
```

```python
# Mostrar todas las filas con valores nulos en la columna Marital_Status
nulos_Marital_Status = df[df['Marital_Status'].isnull()]
print("Filas con valores nulos en 'Marital_Status':")
print(nulos_Marital_Status)

# Contar el número de valores nulos en la columna Marital_Status
nulos_Marital_Status_count = df['Marital_Status'].isnull().sum()
print(f"\nNúmero de valores nulos en 'Marital_Status': {nulos_Marital_Status_count}")
```

```python
moda_scores = df['Marital_Status'].mode()[0]

# Imputar valores nulos con la moda
df['Marital_Status'].fillna(moda_scores, inplace=True)

print(df)
```

```python
frecuencia_marital_status = df['Marital_Status'].value_counts()
print("\nFrecuencia de cada valor único en la columna 'Marital_Status':")
print(frecuencia_marital_status)
```
