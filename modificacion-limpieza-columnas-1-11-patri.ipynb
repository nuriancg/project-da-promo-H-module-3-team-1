```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
df= pd.read_csv('HR RAW DATA.csv')
df.head()
```

### Modificación columna 'Unnamed: 0'

Eliminamos la columna

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
df.head()
```

### Creamos columna 'Age Group'

```python
df.insert(1, 'Age_Group', '')
df.head()
```

### Modificación columna 'Age':

```python
print(df['Age'].dtype)
print(df['Age'].unique())
```

Cambiamos valores con letras a números y cambio dtype 'object' a 'int'

```python
reemplazos = {
    'forty-seven': 47,
    'fifty-eight': 58,
    'thirty-six': 36,
    'fifty-five': 55,
    'fifty-two': 52,
    'thirty-one': 31,
    'thirty': 30,
    'twenty-six': 26,
    'thirty-seven': 37,
    'thirty-two': 32,
    'twenty-four': 24
}

df['Age'] = df['Age'].replace(reemplazos)
```

Comprobamos cambios

```python
df['Age'].unique()
```

Cambiamos tipo de columna de 'object' a 'int'

```python
df['Age'] = df['Age'].astype(int)
print(df['Age'].dtypes)
```

### Modificación columna 'Attrition':

Vemos características de la columna:

```python
print(df['Attrition'].dtype)
print(df['Attrition'].shape)
print(df['Attrition'].unique())
print(df['Attrition'].isnull().sum())
```

Reemplazamos 'dtype' a 'bool'

```python
df['Attrition'] = df['Attrition'].replace({'No': False, 'Yes': True})

print(df['Attrition'].dtype)
df.head()
```

### Modificación columna 'BusinessTravel'

Cambiamos nombre de columna 'BusinessTravel' a 'Business_Travel':

```python
df.rename(columns={'BusinessTravel': 'Business_Travel'}, inplace=True)
df.columns
```

```python
print(df['Business_Travel'].isnull().sum())
print(df['Business_Travel'].notnull().sum())
print(df['Business_Travel'].unique())
```

Reemplazamos guiones (-) por guiones bajos (_):

```python
df['Business_Travel'] = df['Business_Travel'].str.replace('-', '_')
print(df['Business_Travel'].unique())
```

### Modificacion columna 'DailyRate':

```python
print(df['DailyRate'].notnull().sum())
print(df['DailyRate'].isnull().sum())
print(df['DailyRate'].dtype)
```

Cambiamos nombre de columna de 'DailyRate' a 'Daily_Rate':

```python
df.rename(columns={'DailyRate': 'Daily_Rate'}, inplace=True)
df.columns
```

```python
df['Daily_Rate'].unique()
```

Reemplazamos comas por puntos, eliminamos signo de '$' 

```python
df['Daily_Rate'] = df['Daily_Rate'].str.replace(',', '.').str.replace('$', '')
print(df['Daily_Rate'].unique())
```

Convertimos la columna a 'float'

```python
df['Daily_Rate'] = df['Daily_Rate'].astype(float)
print(df['Daily_Rate'].dtype)
```

### Modificación columna 'Department'

```python
print(df['Department'].isnull().sum())
print(df['Department'].notnull().sum())
df['Department'].unique()
```

Hay espacios extras antes y despues de ' Research & Development ', ' Sales ' y  ' Human Resources '

Comprobamos:

```python
rd_count = df['Department'].value_counts().get(['Research & Development', 'Sales', 'Human Resources'], 0)
print(rd_count)
```

```python
rd_count = df['Department'].value_counts().get([' Research & Development ', ' Sales ', ' Human Resources '], 0)
print(rd_count)
```

Eliminamos espacios extras antes y después de los valores usando 'strip()':

```python
df['Department'] = df['Department'].str.strip()
print(df['Department'].unique())
```

### Modificación columna 'DistanceFromHome'

```python
print(df['DistanceFromHome'].dtype)
print(df['DistanceFromHome'].isnull().sum())
print(df['DistanceFromHome'].notnull().sum())
print(df['DistanceFromHome'].unique())
```

Cambiamos Nombre de columna

```python
df.rename(columns={'DistanceFromHome': 'Distance_From_Home'}, inplace=True)
df.columns
```

Existen valores negativos que tenemos que cambiar a positivos. Convertimos de 'int' a 'str' para poder usar método 'replace()':

```python
df['Distance_From_Home'] = df['Distance_From_Home'].astype(str)
df['Distance_From_Home'].dtype
```

Usamos método replace() para transformar valores negativos a positivos.

```python
df['Distance_From_Home'] = df['Distance_From_Home'].str.replace('-', '')
df['Distance_From_Home'].unique()
```

Volvemos a transformar a 'int'

```python
df['Distance_From_Home'] = df['Distance_From_Home'].astype(int)
df['Distance_From_Home'].dtype
```
